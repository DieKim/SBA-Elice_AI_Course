# 방법 1. 각 class의 확률값을 모두 더해 최고값을 가진 class 추출

```python
score = np.sum([score1, score2], axis=0)
score = np.sum([score, score3], axis=0)
score.shape
```

```python
ensemble_1 = [np.argmax(score[i]) for i in range(len(score))]
ensemble_1 = pd.DataFrame({'Emotion' : ensemble_1})
ensemble_1.to_csv('/content/drive/MyDrive/sentimental_analysis_project/ensemble_1.csv', index=False,) # 0.938
```

# 방법 2. Majority voting

```python
score1_list = [np.argmax(score1[i]) for i in range(len(score1))]
score2_list = [np.argmax(score2[i]) for i in range(len(score2))]
score3_list = [np.argmax(score3[i]) for i in range(len(score3))]
```

```python
scores = []

for i in range(len(score1)):
  scores.append([score1_list[i],score2_list[i],score3_list[i]])
  
scores
```

```python
from collections import Counter as cc

score = []

for i in range(len(scores)):
  counter = cc(scores[i])
  mode = counter.most_common(1)
  score.append(mode[0][0])
               
score
```

```python
ensemble_2 = pd.DataFrame({'Emotion' : score})
ensemble_2.to_csv('/content/drive/MyDrive/sentimental_analysis_project/ensemble_2.csv', index=False,) # 0.94
```
